```python
import os
import zipfile

# Define paths
zip_files_directory = os.path.expanduser('~/Work/zip-files')
destination_directory = os.path.expanduser('~/Work/geodata-econ/rpa-landuse-proj/data/spatial-proj')

# Ensure the destination directory exists
os.makedirs(destination_directory, exist_ok=True)

# Function to unzip files
def unzip_files(zip_files_dir, dest_dir):
    for file_name in os.listdir(zip_files_dir):
        if file_name.endswith('.zip'):
            zip_file_path = os.path.join(zip_files_dir, file_name)
            with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
                for member in zip_ref.namelist():
                    # Check if the file is a raster file with no extension or a .hdr file
                    if os.path.splitext(member)[1] == '' or member.endswith('.hdr'):
                        zip_ref.extract(member, dest_dir)

# Unzip the files
unzip_files(zip_files_directory, destination_directory)
```
