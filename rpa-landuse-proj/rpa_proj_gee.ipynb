```python
import ee

# Initialize Earth Engine
ee.Initialize(project='ee-chrismihiar')
```

```python

# Define land use classes
land_use_classes = {
    1: 'Water',
    2: 'Developed',
    4: 'Forest',
    6: 'Other',
    7: 'Pasture',
    8: 'Crop'
}

# Save land use class information to image collections
image_collection = ee.ImageCollection("projects/ee-chrismihiar/assets/2020_RPA_LANDUSE_PROJ/NORESM1_M_RCP85_SSP5")
image_collection = image_collection.set('land_use_classes', land_use_classes)
```

```python
# Get the first image in the collection
first_image = image_collection.first()
first_image.getInfo()
```

```python
import geemap

# Convert the Earth Engine image to an Earth Engine layer
ee_layer = geemap.ee_tile_layer(first_image, {}, 'First Image')

# Create a Map object
Map = geemap.Map()

# Add the Earth Engine layer to the map
Map.addLayer(ee_layer, {}, 'First Image')

# Display the map
Map
```

```python

# Create a visualization
visualization = first_image.visualize(palette=['blue', 'gray', 'green', 'yellow', 'orange', 'red'], min=1, max=8)

# Create a legend
legend = '''
    <h4>Land Use Classes:</h4>
    <ul>
        <li><span style="background-color:blue"></span>Water</li>
        <li><span style="background-color:gray"></span>Developed</li>
        <li><span style="background-color:green"></span>Forest</li>
        <li><span style="background-color:yellow"></span>Other</li>
        <li><span style="background-color:orange"></span>Pasture</li>
        <li><span style="background-color:red"></span>Crop</li>
    </ul>
'''

# Display the visualization and legend
visualization.getInfo()
legend
```
